```python
import requests
import openpyxl
from bs4 import BeautifulSoup

wb=openpyxl.Workbook()
ws=wb.active

aa=['b2b-ph-k-s','b3b-ph-k-s','b4b-ph-k-s','b15b-ph-k-s']

for r in aa:
    url="https://www.digikey.tw/products/zh?keywords="+r
    headers={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.114 Safari/537.36'}
    s=requests.get(url,headers=headers)
    soup=BeautifulSoup(s.text)
    k=soup.select('.productTable')[0].select('tbody')[0].select('tr')[0]
    ws.append([r,\
              k.select('td')[8].span.text.strip(),\
              k.select('td')[6].text.strip()])
wb.save('123.xlsx')
```
