```python
import requests
import time
import datetime
import openpyxl
from bs4 import BeautifulSoup
zz=0
wb =openpyxl.Workbook()
ws=wb.active

ws['A1']='職缺名稱'
ws['B1']='職缺連結'
ws['C1']='公司名稱'
ws['D1']='薪資條件'
ws['E1']='需求經歷'
ws['F1']='需求學歷'
ws['G1']='聯絡人員'


data=requests.get("http://pythonscraping.com/pages/page1.html")
soup= BeautifulSoup(data.text)
page=0
while True:
    page+=1
    data=requests.get("https://www.1111.com.tw/search/job?ks=%E5%A4%A7%E6%95%B8%E6%93%9A&page="+str(page))
    soup= BeautifulSoup(data.text)
    r=soup.select('.srh-body__result')[0].select('li')[0].select('.item__job-info')
    for job in r:
        data2=requests.get(job.div.a['href'])
        soup2= BeautifulSoup(data2.text)
        ws.append([job.div.a.text,\
                  job.div.a['href'],\
                  job.select('div')[2]['aria-label'],\
                  job.select('div')[3].select('i')[1]['aria-label'],\
                  job.select('div')[3].select('i')[2]['aria-label'],\
                  job.select('div')[3].select('i')[3]['aria-label'],\
                  soup2.select('.job-detail-panel--apply-profile-text')[0].select('span')[0].string])
        zz+=1
        print("輸入職缺",zz)
        time.sleep(2.5)
    page=int(page)
    time.sleep(10)
    if page>50:
        break
        
        
        
wb.save('職缺清單'+str(datetime.date.today())+'test.xlsx') 
print("已經成功存檔")
```
