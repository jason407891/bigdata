```python
import requests
from bs4 import BeautifulSoup
data=requests.get("http://pythonscraping.com/pages/page1.html")
soup= BeautifulSoup(data.text)
print(soup)
```

```python
print(soup.prettify())
#可以讀出網頁架構
```

```python
print(soup.select('h1')[0].string)
#soup.select回傳值為陣列 所以如果要取string的話必須加上[0]
```

```python
data=requests.get("https://www.1111.com.tw/search/job?ks=%E5%A4%A7%E6%95%B8%E6%93%9A&fs=1&page=1&ts=1")
soup= BeautifulSoup(data.text)
print(soup)
```

```python
k=soup.select('.srh-body__result')[0].select('li')[0].select('div')[0].select('.item__job-info')[0].div.a.text
print(k)
```

```python
r=soup.select('.srh-body__result')[0].select('li')[0].select('.item__job-info')
for job in r:
    print(job.div.a.text)
```

```python
k=soup.select('.srh-body__result')[0].select('li')[0].select('div')[0].select('.item__job-organ-m')[0]
print(k)
```

```python
import time
r=soup.select('.srh-body__result')[0].select('li')[0].select('.item__job-info')
for job in r:
        print('職缺名稱:'+job.div.a.text)
        print('公司名稱:'+job.select('div')[2]['aria-label'])
        print('薪資條件:'+job.select('div')[3].select('i')[1]['aria-label'])
        print('需求經歷:'+job.select('div')[3].select('i')[2]['aria-label'])
        print('需求學歷:'+job.select('div')[3].select('i')[3]['aria-label'])
        print("")
time.sleep(1)
```
