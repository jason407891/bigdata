```python
import requests
from bs4 import BeautifulSoup

page=0
while True:
    page+=1
    data=requests.get("https://www.1111.com.tw/search/job?ks=%E5%A4%A7%E6%95%B8%E6%93%9A&page="+str(page))
    soup= BeautifulSoup(data.text)
    r=soup.select('.srh-body__result')[0].select('li')[0].select('.item__job-info')
    for job in r:
        print('職缺名稱:'+job.div.a.text)
        print('職缺連結',job.div.a['href'])
        print('公司名稱:'+job.select('div')[2]['aria-label'])
        print('薪資條件:'+job.select('div')[3].select('i')[1]['aria-label'])
        print('需求經歷:'+job.select('div')[3].select('i')[2]['aria-label'])
        print('需求學歷:'+job.select('div')[3].select('i')[3]['aria-label'])
        data2=requests.get(job.div.a['href'])
        soup2= BeautifulSoup(data2.text)
        print("聯絡人:"+soup2.select('.job-detail-panel--apply-profile-text')[0].select('span')[0].string)
        print("")
        time.sleep(1)
    page=int(page)
    if page>3:
        break
```
